```python
from cifar10_emb import Cifar10Emb
from cifar100_emb import Cifar100Emb
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
exp = Cifar10Emb(root_path='.')

exp.dataloader_args['batch_size'] = 32

exp.init_datasets()
# exp.remove_classes()
exp.init_dataloaders()
exp.set_up_model_architecture(300)
exp.init_word_lookup()
```

```python
state, exp_name = (torch.load('./vis_lang/cifar10_bert/models/best_acc.pth'), 'Cifar10 Emb NoCat Fuzzy')
# state, exp_name = (torch.load('./trained_models/vis_lang/cifar10_emb_best_acc.pth')    , 'Cifar10 Emb')
# state, exp_name = (torch.load('./trained_models/vis_lang/cifar10_emb_missing_cat.pth') , 'Cifar10 Emb NoCat')
# state, exp_name = (torch.load('./vis_lang/cifar100_emb/models/best_acc.pth')    , 'Cifar100 Emb')
exp.model.load_state_dict(state['net'])
```

```python
class_embeddings = exp.model.word_lookup.cpu().numpy()
embeddings = [[] for _ in range(len(exp.class_names))]
classes_finished = 0

NUM_EMBEDDINGS_PER_CLASS = 50

for inputs, labels in exp.test_loader:
    with torch.no_grad():
        embedding = exp.model(inputs.cuda()).cpu().numpy()

    for i, label in enumerate(labels):
        if len(embeddings[label]) < NUM_EMBEDDINGS_PER_CLASS:
            embeddings[label].append(embedding[i])
            
            # Done with this class
            if len(embeddings[label]) == NUM_EMBEDDINGS_PER_CLASS:
                classes_finished += 1
    
    if classes_finished >= len(exp.class_names):
        break

# Needed for CIFAR100 where some classes won't be in the validation set
for idx, embedding in enumerate(embeddings):
    while len(embedding) < NUM_EMBEDDINGS_PER_CLASS:
        embedding.append(class_embeddings[idx])
```

```python
words_proj = np.concatenate([class_embeddings] + [np.stack(x, axis=0) for x in embeddings], axis=0)

from sklearn.decomposition import PCA
words_proj = PCA(n_components=50).fit_transform(words_proj)

from sklearn.manifold import TSNE
words_proj = TSNE(n_components=2).fit_transform(words_proj)

print(np.min(words_proj, axis=0), np.max(words_proj, axis=0))
```

```python
class_proj = words_proj[:len(exp.class_names)]
words_proj = words_proj[len(exp.class_names):]

embedding_projs = [None] * len(exp.class_names)

for class_idx in range(len(exp.class_names)):
    embedding_projs[class_idx] = words_proj[:NUM_EMBEDDINGS_PER_CLASS]
    words_proj = words_proj[NUM_EMBEDDINGS_PER_CLASS:]

print(embedding_projs[-1].shape)
```

```python
fig = plt.figure(figsize=(14, 14))
ax = fig.add_subplot(1, 1, 1)

# ax.plot(class_proj[:, 0], class_proj[:, 1], 'ko')
# plt.xlim([-12, 6])
# plt.ylim([-12, 6])

for i, name in enumerate(exp.class_names):
    p = ax.plot(embedding_projs[i][:, 0], embedding_projs[i][:, 1], '.', label=name)
    ax.scatter([class_proj[i, 0]], [class_proj[i, 1]], marker='^', color=p[0].get_color(), s=50)

for i, name in enumerate(exp.class_names):
    ax.annotate(name, (class_proj[i, 0] + 0.1, class_proj[i, 1] + 0.1))

# ax.legend()
plt.title(exp_name + ' Embedding Space')
```
