## When is the forgetting happening?

```python
import json
from matplotlib.pyplot import figure

figure(figsize=(10, 5))


task = 1

with open(f'/nethome/bdevnani3/raid/continual/cifar100_alexnet_base_icl_itl_deep_test_itl/run_0/raw_data_{task}') as f:
  data = json.load(f)
x = range(50)
import matplotlib.pyplot as plt
plt.title(f'Baseline: Accuracies over past tasks training on Task {task}')
for i in range(task+1):
    if i == task:
        linewidth=4.0
    else:
        linewidth=1.0
    plt.plot(x, data["Deep Test Acc"][f"{task}_{i}"], label = f"Task {i} Accuracies", linewidth=linewidth)
plt.xlabel('epochs')
plt.ylabel('accuracies')
plt.legend()
plt.show()


```

```python
import json
from matplotlib.pyplot import figure

figure(figsize=(10, 5))


task = 1

with open(f'/nethome/bdevnani3/raid/continual/cifar100_alexnet_bert_icl_itl_deep_test_itl/run_0/raw_data_{task}') as f:
  data = json.load(f)
x = range(50)
import matplotlib.pyplot as plt
plt.title(f'Bert: Accuracies over past tasks training on Task {task}')
for i in range(task+1):
    if i == task:
        linewidth=4.0
    else:
        linewidth=1.0
    plt.plot(x, data["Deep Test Acc"][f"{task}_{i}"], label = f"Task {i} Accuracies", linewidth=linewidth)
plt.xlabel('epochs')
plt.ylabel('accuracies')
plt.legend()
plt.show()


```
