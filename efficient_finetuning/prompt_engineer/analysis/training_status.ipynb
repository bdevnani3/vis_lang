# Analysis

A collection of functions to quickly analyze different aspects of the training process.

```python
results_path="/nethome/bdevnani3/raid/results/prompt_engineer/"
model_path=  "/nethome/bdevnani3/raid/models/prompt_engineer/"
```

## Training progress

```python
# Check how many epochs are done for each experiment
import os
import json

for d in os.listdir(results_path):
    expt_path = results_path + d +"/plots_raw.json"
    model_expt = model_path + d + "/best_acc.pth"
    if os.path.isfile(expt_path):
#         if os.path.isfile(model_expt):
        if "cifar100_transformer_encoder_100way_1shot" in model_expt:
            data = json.load(open(expt_path))
            n_epochs = len(data["Train Acc"])
            best_train_acc = data["Best Train Acc"]
            print(f"{d}, {n_epochs}/500, {best_train_acc}%")
            print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
```

## Training curves

```python
import matplotlib.pyplot as plt
```

```python
experiment = "cifar100_transformer_encoder_100way_4shot_5episodes_500epochs_lr0.0005_feedforward-dim512_reduction-factor1_kdimNone_vdimNone"
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

def plot_training_curves(experiment):

    expt_path = results_path + "/" + experiment +"/plots_raw.json"
    data = json.load(open(expt_path))

    x= range(len(data["Train Acc"]))
    acc = data["Train Acc"]
    loss = data["Train Loss"]

    fig, ax = plt.subplots(2,1, figsize=(20, 10))
    ax[0].plot(x, acc)
    ax[0].set_title("Training Accuracy")

    ax[1].plot(x, loss)
    ax[1].set_title("Training Loss")

    fig.suptitle(experiment)

    plt.show()
    
plot_training_curves(experiment)
```
