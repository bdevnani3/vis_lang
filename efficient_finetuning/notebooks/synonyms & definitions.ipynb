### 1. Using Wordnet

```python
from nltk.corpus import wordnet as wn
```

```python
def generate_synonym_set(word):
    out = []
    try:
        synset = wn.synset(word+".n.01")
        for lemma in synset.lemmas():
            out.append(lemma.name().lower().replace("_", " "))
        for hypernym in synset.hypernyms():
            out.append(hypernym.name().lower().replace("_", " ").split(".")[0])
        for mh in synset.member_holonyms():
            out.append(mh.name().lower().replace("_", " ").split(".")[0])
        out = list(set(out))
        return out
    except:
        return []

d_synonyms = {}
for c in flowers_classes:
    d_synonyms[c] = generate_synonym_set(c.replace(" ", "_"))
d_synonyms
```

```python
def generate_definition_set(word):
    try:
        synset = wn.synset(word+".n.01")
        return synset.definition()
    except:
        return ""
d_defs = {}
for c in flowers_classes:
    d_defs[c] = generate_definition_set(c.replace(" ", "_"))
d_defs
```

```python
def generate_example_set(word):
    try:
        synset = wn.synset(word+".n.01")
        return synset.examples()
    except:
        return []
d_exs = {}
for c in flowers_classes:
    d_exs[c] = generate_example_set(c)
    print(len(d_exs[c]))
d_exs
```
